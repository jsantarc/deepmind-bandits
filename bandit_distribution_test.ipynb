```python
import numpy as np
import matplotlib.pyplot as plt
from deepmind_bandits import GaussianBandits, UniformBandits, ExponentialBandits, BetaBandits, GammaBandits

# Set random seed for reproducibility
np.random.seed(42)
```

## 1. Gaussian Bandits
Normal distribution with mean and standard deviation parameters.

```python
print("=== Gaussian Bandits ===")
gaussian_bandit = GaussianBandits(means=[1.0, 2.5, 1.8], stds=[0.5, 0.3, 0.8])
print(f"Means: {gaussian_bandit.means}")
print(f"Stds: {gaussian_bandit.stds}")
print(f"Optimal value: {gaussian_bandit.v_star}")
print(f"Pull arm 0: {gaussian_bandit.pull_arm(0):.3f}")
print(f"Pull arm 1: {gaussian_bandit.pull_arm(1):.3f}")
print(f"Pull arm 2: {gaussian_bandit.pull_arm(2):.3f}")
```

```python
# Plot parametric distributions
gaussian_bandit.plot_parametric_distributions(fill=True)
```

```python
# Plot time series with distribution
gaussian_bandit.plot_time_series_with_distribution(T=200, seed=42)
```

## 2. Uniform Bandits

```python
print("=== Uniform Bandits ===")
uniform_bandit = UniformBandits(lows=[0, 1, 2], highs=[2, 4, 5])
print(f"Lows: {uniform_bandit.lows}")
print(f"Highs: {uniform_bandit.highs}")
print(f"Means: {uniform_bandit.means}")
print(f"Optimal value: {uniform_bandit.v_star}")
print(f"Pull arm 0: {uniform_bandit.pull_arm(0):.3f}")
print(f"Pull arm 1: {uniform_bandit.pull_arm(1):.3f}")
print(f"Pull arm 2: {uniform_bandit.pull_arm(2):.3f}")
```

```python
# Plot parametric distributions
uniform_bandit.plot_parametric_distributions(fill=True)
```

```python
# Plot time series with distribution
uniform_bandit.plot_time_series_with_distribution(T=200, seed=42)
```

## 6. Multiple Pulls Test
Test multiple pulls from the same arm to see reward variation.

```python
print("=== Testing Multiple Pulls ===")
np.random.seed(42)
print("\nGaussian Bandit - Arm 1 (mean=2.5, std=0.3):")
for i in range(5):
    reward = gaussian_bandit.pull_arm(1)
    print(f"  Pull {i+1}: {reward:.3f}")

print("\nUniform Bandit - Arm 2 (low=2, high=5):")
for i in range(5):
    reward = uniform_bandit.pull_arm(2)
    print(f"  Pull {i+1}: {reward:.3f}")

print("\nBeta Bandit - Arm 1 (alpha=5, beta=2):")
for i in range(5):
    reward = beta_bandit.pull_arm(1)
    print(f"  Pull {i+1}: {reward:.3f}")
```

## 7. Compare All Distributions
Visualize all distributions side by side.

```python
# Create a figure with subplots for each distribution
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Gaussian
gaussian_bandit.plot_parametric_distributions(ax=axes[0, 0], fill=True, show=False)
axes[0, 0].set_title("Gaussian Bandits", fontweight='bold')

# Uniform
uniform_bandit.plot_parametric_distributions(ax=axes[0, 1], fill=True, show=False)
axes[0, 1].set_title("Uniform Bandits", fontweight='bold')

# Exponential
exp_bandit.plot_parametric_distributions(ax=axes[0, 2], fill=True, show=False)
axes[0, 2].set_title("Exponential Bandits", fontweight='bold')

# Beta
beta_bandit.plot_parametric_distributions(ax=axes[1, 0], fill=True, show=False)
axes[1, 0].set_title("Beta Bandits", fontweight='bold')

# Gamma
gamma_bandit.plot_parametric_distributions(ax=axes[1, 1], fill=True, show=False)
axes[1, 1].set_title("Gamma Bandits", fontweight='bold')

# Hide the last subplot
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

## 8. Empirical Mean Verification
Verify that sample means converge to theoretical means.

```python
np.random.seed(42)
num_samples = 10000

print(f"=== Empirical Mean Verification ({num_samples} samples) ===")

# Gaussian
print("\nGaussian Bandit:")
for arm in range(gaussian_bandit.num_arms):
    samples = [gaussian_bandit.pull_arm(arm) for _ in range(num_samples)]
    empirical_mean = np.mean(samples)
    theoretical_mean = gaussian_bandit.means[arm]
    print(f"  Arm {arm}: Theoretical={theoretical_mean:.3f}, Empirical={empirical_mean:.3f}")

# Uniform
print("\nUniform Bandit:")
for arm in range(uniform_bandit.num_arms):
    samples = [uniform_bandit.pull_arm(arm) for _ in range(num_samples)]
    empirical_mean = np.mean(samples)
    theoretical_mean = uniform_bandit.means[arm]
    print(f"  Arm {arm}: Theoretical={theoretical_mean:.3f}, Empirical={empirical_mean:.3f}")

# Beta
print("\nBeta Bandit:")
for arm in range(beta_bandit.num_arms):
    samples = [beta_bandit.pull_arm(arm) for _ in range(num_samples)]
    empirical_mean = np.mean(samples)
    theoretical_mean = beta_bandit.means[arm]
    print(f"  Arm {arm}: Theoretical={theoretical_mean:.3f}, Empirical={empirical_mean:.3f}")
```
